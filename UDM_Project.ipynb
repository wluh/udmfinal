```python
df = pd.read_csv(r'trees.csv')
```

```python
def get_season(input):
  return input['PlntSeas'][:-2]

def get_year(input):
  return int(input['PlntSeas'][-2:]) + 2000
```

```python
df.head()
```

```python
df['SEASON'] = df.apply(lambda row: get_season(row), axis=1)
df['YEAR'] = df.apply(lambda row: get_year(row), axis=1)
```
